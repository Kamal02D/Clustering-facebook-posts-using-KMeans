```python
# Dropping the empty columns
```

```python
df = df[df.columns[:12]]
df
```

let's check now if there is any null values in some lignes

```python
for i in range(df.shape[0]):
    row = df.iloc[i,:]
    for column_name in df.columns:
        if row[column_name] is None:
            print(i)
```

now we will drop the status_published date because it have no impact in our clustering

```python
df = df[list(df.columns[:2])+list(df.columns[3:])]
df
```

Now we will transform the status_type to numerical values

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(df["status_type"])
```

```python
df["status_type"] = le.transform(df["status_type"])
```

```python
df=df[df.columns[1:]] # deleting the status id
df.head(1)
```

# importing the model

```python
from sklearn.cluster import KMeans
```

# finding the best number of clusters using the elbow method

```python
inertia_s = []
for i in range(1,20):
    print(f"itteration {i} out of {df.shape[0]}")
    kmeans = KMeans(n_clusters=i,n_init=15)
    kmeans.fit(df)
    inertia_s.append(kmeans.inertia_) # the sum of the square distances of all the clusters
```

Plotting the inertia_ of every number of clusters

```python
import matplotlib.pyplot as plt
plt.plot(range(1,20) , inertia_s)
```

it looks like that the best k could be 2 , 3

```python
kmeans = KMeans(n_clusters=2 , n_init=15)
```

```python
kmeans.fit(df)
```

```python
df["class"]  =kmeans.labels_
```

```python
df.value_counts("class")
```

# analyzing the result

```python
for i in range(2):
    print(df[df["class"] == i].describe())
```

it looks like class 0 is for the best engaging post and 1 for the less engaging

```python
condition_ = lambda class_: "good" if class_ == 0 else "bad"
```

```python
df["class"] = df["class"].apply(condition_)
```
